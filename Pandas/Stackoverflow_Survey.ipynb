```python
country_grp = df.groupby(['Country'])
```

```python
country_grp.get_group('India')
```

```python
country_grp['Age'].value_counts().loc['India']
```

```python
country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# Percetage of users in a country with python
country_total_responses = df['Country'].value_counts()
country_total_responses
```

```python
country_python_users = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_python_users
```

```python
percentage_python_country = country_python_users / country_total_responses * 100
percentage_python_country
```

```python
python_df = pd.concat([country_total_responses, country_python_users], axis='columns', sort=False)
```

```python
python_df.rename(columns={'count': 'Responses', 'LanguageHaveWorkedWith': 'Knows Python'}, inplace=True)
```

```python
python_df['PercentKnowsPython'] = (python_df['Knows Python'] / python_df['Responses']) * 100
```

```python
python_df.sort_values(by='PercentKnowsPython', ascending=False, inplace=True)
```
